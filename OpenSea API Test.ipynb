# Getting all the collections of a specific wallet owner

```python
wallet_address = "0x696c53f2c21769ed51a6b6514d25763422980551"
```

```python
url = "https://api.opensea.io/api/v1/collections"
querystring = {"asset_owner":wallet_address,"limit":"300","offset":"0"}
response = requests.request("GET", url, params=querystring)
response_json = response.json()
response_json
```

```python
wallet_collections = {}
df_columns = ['Collection Slug', 'Image URL', 'Total Supply', 'Number of Owners', 'Total Volume', 'Market Cap', 'Floor Price', 'Average Price',
          'One Day Volume', 'One Day Change', 'One Day Sales', 'One Day Average Price',
          'Seven Day Volume', 'Seven Day Change', 'Seven Day Sales', 'Seven Day Average Price',
          'Thirty Day Volume', 'Thirty Day Change', 'Thirty Day Sales', 'Thirty Day Average Price'
          ]

for collection in response_json:
    wallet_collections[collection['name']] = [
        collection['slug'],
        collection['image_url'],        
        collection['stats']['total_supply'],
        collection['stats']['num_owners'],
        collection['stats']['total_volume'],
        collection['stats']['market_cap'],
        collection['stats']['floor_price'],
        collection['stats']['average_price'],
        collection['stats']['one_day_volume'],
        collection['stats']['one_day_change'],
        collection['stats']['one_day_sales'],
        collection['stats']['one_day_average_price'],
        collection['stats']['seven_day_volume'],
        collection['stats']['seven_day_change'],
        collection['stats']['seven_day_sales'],
        collection['stats']['seven_day_average_price'],
        collection['stats']['thirty_day_volume'],
        collection['stats']['thirty_day_change'],
        collection['stats']['thirty_day_sales'],
        collection['stats']['thirty_day_average_price']
    ]
```

# Displaying items in tabular grid

```python
labels = []
token_ids = []
images = []

for collection in wallet_collections:
    
    slug = wallet_collections[collection][0]  
    url = "https://api.opensea.io/api/v1/assets"
    querystring = {"owner":"0x696c53f2c21769ed51a6b6514d25763422980551","collection":f"{slug}","limit":"50","offset":"0"}
    response = requests.request("GET", url, params=querystring)
    collection = response.json()
    
    i = 0
    for item in collection['assets']:
        
        labels.append(collection['assets'][i]['collection']['name'])
        token_ids.append(item['token_id'])
        
        img_name = collection['assets'][i]['collection']['name']+".png"
        img_url = ""
                            
        if "opensea" in collection['assets'][i]['image_url']:
            img_url = "https://lh3.googleusercontent.com/YEs1s94Pxj5MmnruPXJ1p_KJPyd-oUEmLYeMCRwov1C59Ows3yWuzPh4AnS7K2XIdj0bZgMIVKzWKyM2CUy8DO4uYIMwEs4gb7a1Ew=s120"
        elif collection['assets'][i]['image_url'] is None:
            img_url = "https://lh3.googleusercontent.com/YEs1s94Pxj5MmnruPXJ1p_KJPyd-oUEmLYeMCRwov1C59Ows3yWuzPh4AnS7K2XIdj0bZgMIVKzWKyM2CUy8DO4uYIMwEs4gb7a1Ew=s120"
        else:
            img_url = collection['assets'][i]['image_url']
            
        urllib.request.urlretrieve(img_url, img_name)
        img = Image.open(img_name)
    
        images.append(img)
        
        i = i + 1
```

```python
ipyplot.plot_class_tabs(images, labels, custom_texts=token_ids)
```

# Rumble Kong League

```python
api_url = 'https://api.opensea.io/api/v1/assets?collection=rumble-kong-league&limit=50&owner=0x696c53f2c21769ed51a6b6514d25763422980551'
headers =  {"Content-Type":"application/json"}
```

```python
response = requests.get(api_url, headers=headers)
response_json = response.json()
response_json
```

```python
kongs = {}
boosts = {}

for item in response_json['assets']:
    
    for trait in item['traits']:
            if trait['trait_type'] in ('Vision', 'Defense', 'Finish', 'Shooting'):
                boosts[trait['trait_type']] = trait['value']
    
    kongs[item['name']] = [
        item['token_id'],
        item['image_url'],
        boosts
    ]
     
print(kongs)
```

```python
labels = []
images = []
img_name = ""
img_url = ""

for kong in kongs:
    
    labels.append(kong)
    
    img_name = f'{kong}.png'
    
    img_url = kongs[f'{kong}'][1]
        
    urllib.request.urlretrieve(img_url, img_name)
    img = Image.open(img_name)
    
    images.append(img)
```

```python
ipyplot.plot_images(images, labels, img_width=200)
```

```python
fig, axs = plt.subplots(int(np.ceil(len(images))/2), int(np.ceil(len(images))/2), figsize=(10,10))

img = 0

for i in range(0, int(np.ceil(len(images)/2))):
    
    for j in range(0, int(np.ceil(len(images)/2))):    
        
        axs[i][j].imshow(images[img])
        axs[i][j].set_title(labels[img])
        axs[i][j].axis('off')
        
        img = img + 1
```
